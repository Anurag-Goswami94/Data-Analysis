# Data Analysis and Visualizations

### Research Question-1

```python
cancelled_perc= df_hotel['is_canceled'].value_counts(normalize=True) ## Will return the percentage of the unique values
print (cancelled_perc)
```

```python
plt.figure(figsize= (5,6))
plt.title('Reservation Status Count.')
plt.bar(['Not Canceled', 'Canceled'], df_hotel['is_canceled'].value_counts())
plt.show()
```

### Depending on the 2 hotels whose cancelation rate is more:

```python
plt.figure(figsize=(9,9))
ax1=sns.countplot(x= 'hotel', hue= 'is_canceled', data= df_hotel, palette= 'Blues')
plt.title('Reservation status in different hotels.', size=20)
plt.xlabel('Hotel.')
plt.ylabel('Number Of Reservations.')
plt.legend(('not canceled', "canceled"))
plt.show()
```

```python
resort_hotel= df_hotel[df_hotel['hotel']== 'Resort Hotel'] ## Filtering the resort hotel
resort_hotel['is_canceled'].value_counts(normalize= True)
```

```python
city_hotel= df_hotel[df_hotel['hotel']== 'City Hotel'] ## Filtering the city hotel
city_hotel['is_canceled'].value_counts(normalize= True)
```

### Checking if price have any affect on the cancelation of city and resort hotel.

```python
resort_hotel= resort_hotel.groupby('reservation_status_date') [['adr']].mean()
city_hotel= city_hotel.groupby('reservation_status_date') [['adr']].mean() 
```

### Creating visualization for the Average Daily Rate (ADR)

```python
plt.figure(figsize= (20,8))
plt.title ('ADR in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')## we used '.index' to convert the Date-Time-Index as an Index. Can also do it without making it an index
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')## we used '.index' to convert the Date-Time-Index as an Index. Can also do it without making it an index
plt.legend(fontsize= 20)
plt.show()
```

### Checking in which months there is more reservation and which month there is more cancelation.

```python
df_hotel['month']= df_hotel['reservation_status_date'].dt.month ## This line will convert all the Date-Time-Index by Months creating a "Month" column in the hotel dataframe.
plt.figure(figsize= (16, 8))
ax1= sns.countplot(x= 'month', hue= 'is_canceled', data= df_hotel, palette= 'bright')
plt.title('Reservation Status Per Month', size=20)
plt.xlabel('Month')
plt.ylabel('Number Of Reservation')
plt.legend(['Not Canceled', 'Canceled'])
plt.show()
```

### Plotting the ADR for each month to check if price affect the cancelation rate

```python
plt.figure(figsize= (15,8))
plt.title('ADR for each Month', fontsize= 30)
sns.barplot('month', 'adr', data= df_hotel[df_hotel['is_canceled']==1].groupby("month")[['adr']].sum(). reset_index())## We performed the visualisation on only canceled data to see if the price have any affect on the cancelation rates.
plt.show() ## We dont need legend here because we are just checking the price in Monthly basis.
```

### Checking the cancelation on the basis of Countries (Only top 10)

```python
cancelled_data = df_hotel[df_hotel['is_canceled']==1]## Filtering the canceled data
top_10_countries = cancelled_data['country'].value_counts()[:10] ## Counting the top 10 unique values in the country colm. 
plt.figure(figsize=(10,10))
plt.title('Top 10 countries with cancelled reservation')
plt.pie(top_10_countries, autopct = '%.2f',labels = top_10_countries.index)
plt.show()
```

### Checking if the ADR was high during cancelation or during non cancelation

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace= True)

not_cancelled_data= df_hotel[df_hotel['is_canceled']==0]## Filtering the non-canceled data

not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace= True)

plt.figure(figsize=(20,8))
plt.title('Average Daily Rate', fontsize=30)
plt.plot (not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not cancelled')
plt.plot (cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
plt.legend(fontsize=10)
plt.show()
```

#### Filtering data of ADR from 2016 to September-2017 because for the other years the date is not consicutive.

```python
cancelled_df_adr= cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr= not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate', fontsize=30)
plt.plot (not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not cancelled')
plt.plot (cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
plt.legend(fontsize=10)
plt.show()
```
